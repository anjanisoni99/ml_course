```python
def Salary_clean(salary):
    if isinstance(salary,str):
        salary= salary.strip()
        if '$'in salary:
            salary=salary.replace('$','')
        if ',' in salary:
            salary=salary.replace(',','')
        return int (salary)
    return salary
```

```python
df['Salary'].apply(Salary_clean)
```

```python
df['Salary'] = df['Salary'].apply(Salary_clean)
df
```
